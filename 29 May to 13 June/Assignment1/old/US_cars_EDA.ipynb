# **UPLOADING DATA**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('USA_cars_datasets-2.csv')
df.head()
```

# **DATA CLEANING**

removing some insignificant columns

```python
drop_columns=['Unnamed: 0', 'vin', 'lot']
df= df.drop(drop_columns, axis=1)
```

```python
df.head()
```

removing some insignificant rows, such as those having zero prices

```python
drop_rows= df.loc[df['price']==0].index
```

```python
df= df.drop(drop_rows)
```

# **ANALYSIS OF DATA FROM THE PERSPECTIVE OF PRICE**

```python
import matplotlib.pyplot as plt
```

plotting a bar graph of price vs year

```python
price = df['price']
year = df['year']
plt.figure(figsize=(6, 6))
plt.bar(year,price)
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Year vs Price')
plt.show()
```

plotting a bar graph of brand vs price

```python
price = df['price']
year = df['brand']
plt.figure(figsize=(20, 6))
plt.bar(year,price)
plt.xlabel('brand')
plt.ylabel('price')
plt.title('Brand vs Price')
plt.show()
```

plotting price vs mileage

```python
price = df['price']
mileage = df['mileage']
plt.figure(figsize=(15, 6))
plt.plot(mileage,price)
plt.xlabel('mileage')
plt.ylabel('price')
plt.title('mileage vs Price')
plt.show()
```

plotting title status vs price

```python
price = df['price']
status = df['title_status']
plt.figure(figsize=(3, 6))
plt.bar(status,price)
plt.xlabel('status')
plt.ylabel('price')
plt.title('status vs Price')
plt.show()
```

```python
price = df['price']
color = df['color']
plt.figure(figsize=(16, 6))
plt.bar(color,price)
plt.xlabel('color')
plt.ylabel('price')
plt.title('color vs Price')
plt.show()
```

```python
price = df['price']
state = df['state']
plt.figure(figsize=(20, 6))
plt.bar(state,price)
plt.xlabel('state')
plt.ylabel('price')
plt.title('state vs Price')
plt.show()
```

```python
price = df['price']
country = df['country']
plt.figure(figsize=(5, 6))
plt.bar(country,price)
plt.xlabel('country')
plt.ylabel('price')
plt.title('country vs Price')
plt.show()
```

# **CONCLUSIONS :**


*   **Year** : It is evident that there is a clear trend where the price of a car tends to decrease as it gets older.
*   **Brand** : As depicted in the graph, there is a significant variation in price depending on the car brand.
*   **Mileage** : There is no clear relation between price and mileage. It will create noise in the machine model.
*   **Title_status** : Clearly, clean vehicles fall in higher price range.
*   **Color** : It is also a depending feature.
*   **State and Country** : They are also relevant variables.
